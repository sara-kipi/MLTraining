```python
def clean_str(string):
    
    string = re.sub(r"\'s", "", string)
    string = re.sub(r"\'ve", "", string)
    string = re.sub(r"n\'t", "", string)
    string = re.sub(r"\'re", "", string)
    string = re.sub(r"\'d", "", string)
    string = re.sub(r"\'ll", "", string)
    string = re.sub(r",", "", string)
    string = re.sub(r"!", " ! ", string)
    string = re.sub(r"\(", "", string)
    string = re.sub(r"\)", "", string)
    string = re.sub(r"\?", "", string)
    string = re.sub(r"'", "", string)
    string = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", string)
    string = re.sub(r"[0-9]\w+|[0-9]","", string)
    string = re.sub(r"\s{2,}", " ", string)
    return string.strip().lower()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
#function to implement lemmatization
def lemma(string):
    
    words = []
    tokens = nltk.word_tokenize(string)
    for word in tokens:
        #pint(word)
        words.append(lemmatizer.lemmatize(word))
        
    text = " ".join(words)
    return text
```

```python
#### Testing the function lemma
text = "Ink helps drive democracy in Asia"
lemma(text)
```

```python
data = data.copy()
data['Lemmatized'] = data['Clean_Text'].apply(lambda x: lemma(x))
```

```python
data.head()
```

#### “PresenceAbsenceVector”: Converts Article to vectors using Presence and Absence of Words

```python
word_tokens = []
for item in data['Lemmatized'].values:
    tok = nltk.word_tokenize(item)
    for t in tok:
        word_tokens.append(t)
review_tokens = set(word_tokens)
```

```python
# Processing Tokens
review_dict = dict.fromkeys(review_tokens,0)
```

```python
len(word_tokens)
```

```python
len(review_dict)
```

```python
#creating the Presence Absence Vector
row = []
for items in data['Lemmatized'].values:
    review_dict = dict.fromkeys(review_tokens,0)
    wt = nltk.word_tokenize(items)
    for tokens in wt:
        review_dict[tokens]+=1
    row.append(review_dict)
```

```python
#create a pandas dataframe
reviews_df = pd.DataFrame(row)
```

```python
reviews_df.head()
```

```python
reviews_df.shape
```

```python
#Applying KMeans Clustering to the Presence Absence Vector
from sklearn.cluster import KMeans
X = reviews_df.values
from sklearn.metrics import silhouette_samples, silhouette_score
clusterer = KMeans(n_clusters=5, random_state=10)
cluster_labels = clusterer.fit_predict(X)
```

```python
len(clusterer.labels_)
```

```python
from collections import Counter
Counter(cluster_labels)
```

```python
#creating a dataframe containing clustering labels for presence absence vector
df_presence= pd.DataFrame({'Cluster':clusterer.labels_})
```

```python
df_presence.head()
```

```python
# The silhouette_score gives the average value for all the samples.
# This gives a perspective into the density and separation of the formed
# clusters
n_clusters =5
vect =[]
score = []
silhouette_avg = silhouette_score(X, cluster_labels)
print("For n_clusters =", n_clusters,"The average silhouette_score is :", silhouette_avg)
vect.append('PresenceAbsense')
score.append(silhouette_avg)
# Compute the silhouette scores for each sample
sample_silhouette_values = silhouette_samples(X, cluster_labels)
```

```python
type(X)
```

```python
#dimensionality reduction
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=2,random_state=42)
data_reduced = svd.fit_transform(X)
```

```python
data_reduced = pd.DataFrame(data_reduced)
```

```python
#Visualization of PresenceAbsence Vector
ax = data_reduced.plot(kind='scatter', x=0, y=1, c=cluster_labels, cmap='rainbow',figsize=(15,8))
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('Projection of the clustering on a the axis of TruncatedSVD on PresenceAbsence Vector')

for x, y, label in zip(data_reduced[0], data_reduced[1], clusterer.labels_):
    ax.annotate('C{0}'.format(label), (x,y))
```

#### What does Silhouette Coefficienttell us?

An overall measure of the goodness of a clustering can be obtained by computing the average silhouette coefficient of all points.

The silhouette value is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation). The silhouette ranges from −1 to +1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. If most objects have a high value, then the clustering configuration is appropriate. If many points have a low or negative value, then the clustering configuration may have too many or too few clusters.

# Text Summarization :
    
    Process of shortening a text document or enormous amount of textual material like web pages blogs etc

```python
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
#import dataset
with open('President-Speech.txt','r') as f:
          dataset = f.read()
```

```python
dataset
```

```python
def summarize(document):
    result = []
    for number,sentence in enumerate(sent_tokenize(document)):
        number_tokens = len(word_tokenize(sentence))
        tagged = nltk.pos_tag(word_tokenize(sentence))
        number_nouns = len([word for word,pos in tagged if pos in ["NN","NNP"]])
        #named entity recognition
        ners = nltk.ne_chunk(nltk.pos_tag(word_tokenize(sentence)),binary=False)
        number_ners = len([chunk for chunk in ners if hasattr(chunk,'label') ])
        #higher the score higher the importance
        score = (number_ners + number_nouns)/float(number_tokens)
        result.append((number,score,sentence))
    return result
```

```python
summ = summarize(dataset)
```

```python
summ
```

```python
## Print the Summary ##
#sentence with higher score are above
#the result is sorted on descending order of score
for i in sorted(summ, key=lambda x: x[1], reverse=True):
    print('='*20)
    print('Score : ',i[1] )
    print('='*20)
    print('Summary : ',i[2] )
```

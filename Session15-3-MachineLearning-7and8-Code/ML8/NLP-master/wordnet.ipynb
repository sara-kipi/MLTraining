# Movie Review Classification -Text Classification

```python
import nltk
import random
from nltk.corpus import movie_reviews
```

```python
#create a list of documents
documents = []
```

```python
movie_reviews.fileids('neg')[0]
```

```python
for categories in movie_reviews.categories():
    for fileid in movie_reviews.fileids(categories):
        documents.append((list(movie_reviews.words(fileid)),categories))
```

```python
print(documents[1])
```

```python
#shuffle the documents
random.shuffle(documents)
```

```python
print(documents[1])
```

```python
#normalize the dataset 
all_words = []
for words in movie_reviews.words():
    all_words.append(words.lower())
```

```python
#convert word list into nltk frequency distribution
all_words = nltk.FreqDist(all_words)
print(all_words.most_common(15))
print(all_words["love"])
```

```python
#limit the words
word_features = list(all_words.keys())[:3000]
```

```python
len(word_features)
```

```python
word_features[1]
```

```python
#find features within the document 
def find_features(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features
```

```python
print(find_features('neg/cv000_29416.txt'))
```

```python
for rev,category in documents:
    print(rev,category)
    break
```

```python
feature_set = [ (find_features(rev),category) for (rev,category) in documents]
```

```python
feature_set[0]
```

```python
#split the feature set into training an test set
training_set = feature_set[:1900]
test_set = feature_set[1900:]
```

```python
#train the model using nltk.NaiveBayes Classifier
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
#finding the accuracu score
accuracy = nltk.classify.accuracy(classifier,test_set)* 100
```

```python
print(accuracy)
```

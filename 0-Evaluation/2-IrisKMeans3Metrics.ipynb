```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score


# Load the Iris dataset
iris = load_iris()
X = iris.data
df = pd.DataFrame(X, columns=iris.feature_names)
df
```

```python

# Perform K-means clustering with the optimal number of clusters (3)
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
labels = kmeans.labels_
df
```

```python
# Add the actual target labels for comparison
df['Actual'] = iris.target
df
```

```python
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(X, labels)
print(f'Silhouette Score: {sil_score}')
```

Output:
The silhouette score is printed, indicating the quality of the clustering. A higher score closer to 1 means better-defined clusters, while a score closer to -1 indicates that the clusters are not well-defined.

```python
ch_score = calinski_harabasz_score(X, labels)
print(f'Calinski-Harabasz Index: {ch_score}')
```

```python
db_score = davies_bouldin_score(X, labels)
print(f'Davies-Bouldin Index: {db_score}')
```
